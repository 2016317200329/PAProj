# 1. Calculating trial for ascending-price in symmetry situation
1. `b` is bidfee,`s` is bidinc,`v` is valuation. And for ascending-price, there is $n \leq Q == \lfloor (v-b/s) \rfloor$
2. 公式1如下:
$$ u[i] = 1-b/(v-s*(i-1)),wrt\ i \geq 1\ and\ i \leq Q $$
3. According to Chain Rule, there is Eq2:
$$ p(Duration(or\ 'n') = i) = (1-u[i+1])*u[i]*[i-1]..*u[1] $$
4. `u`可以利用setting值计算，计算`p`之前必须先计算出`u`

Setting设置：`b` is bidfee,`s` is bidinc,`v` is valuation.

```python
b = 0.6
s = 0.12
v = 30
```

计算`Q`. For ascending-price, there is $n \leq Q $

```python
import numpy as np

Q = np.int(np.floor((v-b)/s))
#TODO： 合理性判断的assert
print("Q is : ", Q)
```

计算`u`，注意边界值取值问题：根据公式可知`u[0]`不存在，或者人为指定一个值？

```python
u=[0]*(Q+1)
u[0] = 0
for i in range(1,Q+1):
    u[i] = 1-b/(v-s*(i-1))

print("u is like: ", u)
```

计算`p`，注意起始位置是0,而且p[i] = (1-u[i+1])*u[i]*[i-1]..*u[1]

```python
P=[0]*(Q)
P[0] = (1-u[1])
tmp = 1
for i in range(1,Q):
    tmp = tmp*u[i]
    P[i] = (1-u[i+1])*tmp

print("P is like: ", P)
print("And sum of P is:", np.sum(P))
```

5. 因此`P[n]`表示的是auction结束在第`n`轮的概率, 注意拍卖有可能结束在第`0`轮
6. 注意P的概率大小差别很大（8个数量级）
